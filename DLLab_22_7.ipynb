```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.datasets import load_iris
```

```python
df = load_iris()
```

```python
load_iris()
```

```python
df.feature_names
```

```python
df.target_names
```

```python
X = df.data
y = df.target
```

```python
target_names = df.target_names
```

```python
#petal length is column 2, petal width is column 3
x_feature = 2
y_feature = 3

#markers for each column
markers = ['o', 's', '^']
colors = ['red', 'green', 'blue']
```

```python
plt.figure(figsize=(6,4))
for i, label in enumerate(np.unique(y)):
  plt.scatter(X[y == label, x_feature],
              X[y == label, y_feature],
              marker = markers[i],
              color = colors[i],
              label = target_names[label],
              edgecolor = 'black')

plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')
plt.title('Petal length vs Petal width')
plt.legend()
plt.grid(True)
plt.tight_layout()
```
